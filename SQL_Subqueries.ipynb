## Subqueries

### 1️⃣ Write subquery in WHERE clause.
Find employees who earn more than the company average salary.

```python
SELECT name
FROM employees
WHERE salary > (SELECT AVG(salary) FROM employees);
```

### 2️⃣ Write correlated subquery.
Find employees who earn more than the average salary of their own department.

```python
SELECT e.name, e.department, e.salary
FROM employees e
WHERE e.salary > (
    SELECT AVG(e2.salary)
    FROM employees e2
    WHERE e2.department = e.department
);
```

### 3️⃣ Find second highest salary.
Write a query to get the second highest distinct salary.

```python
SELECT DISTINCT salary
FROM employees
ORDER BY salary DESC
LIMIT 1 OFFSET 1;
```

### 4️⃣ Find nth highest salary.
Write a query to get the 3rd highest distinct salary (N = 3).

```python
SELECT DISTINCT salary
FROM employees
ORDER BY salary DESC
LIMIT 1 OFFSET 2;
```

### 5️⃣ Employees earning more than average salary.
Use a subquery in WHERE clause.

```python
SELECT name
FROM employees
WHERE salary > (SELECT AVG(salary) FROM employees);
```

### 6️⃣ EXISTS vs IN.
Example using IN and EXISTS.

```python
-- Using IN
SELECT name
FROM employees
WHERE department IN (SELECT department FROM departments);

-- Using EXISTS
SELECT e.name
FROM employees e
WHERE EXISTS (
    SELECT 1
    FROM departments d
    WHERE d.department = e.department
);
```

### 7️⃣ Scalar subquery example.
Display employee name and company average salary as a column.

```python
SELECT name,
       (SELECT AVG(salary) FROM employees) AS company_avg_salary
FROM employees;
```

### 8️⃣ Subquery vs JOIN difference.
Example comparison.

```python
-- Using Subquery
SELECT name
FROM employees
WHERE department IN (
    SELECT department FROM departments
);

-- Using JOIN
SELECT e.name
FROM employees e
JOIN departments d
ON e.department = d.department;
```
