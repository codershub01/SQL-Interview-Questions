# SQL CASE & Conditional Logic Practice (With Solutions and Explanation)

## Example Table: employees

We will use this table for all examples.


```python
CREATE TABLE employees (
    emp_id INT,
    name VARCHAR(50),
    department VARCHAR(50),
    salary INT
);

INSERT INTO employees VALUES
(1,'Alice','HR',50000),
(2,'Bob','IT',70000),
(3,'Charlie','IT',80000),
(4,'David','Finance',60000),
(5,'Emma','HR',75000),
(6,'Frank','IT',80000),
(7,'Grace','HR',50000);
```

## 1. CASE Statement (Categorize Salary)

**Goal:** Categorize employee salaries into High, Medium, Low.

Rules:

- salary >= 75000 → High  
- 60000–74999 → Medium  
- <60000 → Low


```python
SELECT name,
salary,
CASE
    WHEN salary >= 75000 THEN 'High'
    WHEN salary BETWEEN 60000 AND 74999 THEN 'Medium'
    ELSE 'Low'
END AS salary_category
FROM employees;
```

### Explanation

CASE works like IF-ELSE in SQL.

It evaluates conditions sequentially and returns the first matching result.


### Example Output

|name|salary|salary_category|
|----|----|----|
|Alice|50000|Low|
|Bob|70000|Medium|
|Charlie|80000|High|


## 2. Conditional Aggregation

**Goal:** Count employees in each salary category.


```python
SELECT
COUNT(CASE WHEN salary < 60000 THEN 1 END) AS low_salary,
COUNT(CASE WHEN salary BETWEEN 60000 AND 74999 THEN 1 END) AS medium_salary,
COUNT(CASE WHEN salary >= 75000 THEN 1 END) AS high_salary
FROM employees;
```

### Explanation

Conditional aggregation counts rows that satisfy specific conditions.

Pattern used:

COUNT(CASE WHEN condition THEN 1 END)


### Example Output

|low_salary|medium_salary|high_salary|
|----|----|----|
|2|2|3|


## 3. Pivot Rows Into Columns

**Goal:** Count employees per department.


```python
SELECT
SUM(CASE WHEN department = 'HR' THEN 1 ELSE 0 END) AS HR,
SUM(CASE WHEN department = 'IT' THEN 1 ELSE 0 END) AS IT,
SUM(CASE WHEN department = 'Finance' THEN 1 ELSE 0 END) AS Finance
FROM employees;
```

### Explanation

This technique is called manual pivoting.

Each department becomes its own column using CASE conditions.


### Example Output

|HR|IT|Finance|
|--|--|--|
|3|3|1|


## 4. Conditional Aggregation with GROUP BY

**Goal:** Count salary categories per department.


```python
SELECT
department,
SUM(CASE WHEN salary >= 75000 THEN 1 ELSE 0 END) AS high_salary,
SUM(CASE WHEN salary BETWEEN 60000 AND 74999 THEN 1 ELSE 0 END) AS medium_salary,
SUM(CASE WHEN salary < 60000 THEN 1 ELSE 0 END) AS low_salary
FROM employees
GROUP BY department;
```

### Explanation

This combines CASE for classification, SUM for counting rows, and GROUP BY to group results per department.


### Example Output

|department|high_salary|medium_salary|low_salary|
|--|--|--|--|
|HR|1|0|2|
|IT|2|1|0|
|Finance|0|1|0|
