# SQL Practice Notebook

## Basic SELECT Queries (With Solutions)

### 1️⃣ Select all columns from a table.

```python
SELECT * FROM employees;
```

### 2️⃣ Select specific columns.

```python
SELECT name, salary FROM employees;
```

### 3️⃣ Filter rows using WHERE.

```python
SELECT * FROM employees WHERE department = 'IT';
```

### 4️⃣ Use AND, OR, NOT.

```python
SELECT * FROM employees WHERE department = 'IT' AND salary > 75000;
```

### 5️⃣ Use IN and NOT IN.

```python
SELECT * FROM employees WHERE department IN ('HR', 'Finance');

SELECT * FROM employees WHERE department NOT IN ('IT');
```

### 6️⃣ Use BETWEEN.

```python
SELECT * FROM employees WHERE salary BETWEEN 60000 AND 75000;
```

### 7️⃣ Use LIKE with wildcards.

```python
SELECT * FROM employees WHERE name LIKE 'A%';
```

### 8️⃣ Find NULL values.

```python
SELECT * FROM employees WHERE email IS NULL;
```

### 9️⃣ Replace NULL using COALESCE.

```python
SELECT name, COALESCE(bonus, 0) AS bonus FROM employees;
```

### 🔟 Sort results using ORDER BY.

```python
SELECT * FROM employees ORDER BY salary DESC;
```
