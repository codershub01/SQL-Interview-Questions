# SQL Window Functions Practice (With Solutions + Outputs)

## Example Table: employees

```python
CREATE TABLE employees (
    emp_id INT,
    name VARCHAR(50),
    department VARCHAR(50),
    salary INT
);

INSERT INTO employees VALUES
(1,'Alice','HR',50000),
(2,'Bob','IT',70000),
(3,'Charlie','IT',80000),
(4,'David','Finance',60000),
(5,'Emma','HR',75000),
(6,'Frank','IT',80000),
(7,'Grace','HR',50000);
```

### Table Data

| emp_id | name | department | salary |
|------|------|------|------|
|1|Alice|HR|50000|
|2|Bob|IT|70000|
|3|Charlie|IT|80000|
|4|David|Finance|60000|
|5|Emma|HR|75000|
|6|Frank|IT|80000|
|7|Grace|HR|50000|


## 1. ROW_NUMBER() – Unique ranking by salary

```python
SELECT *,
ROW_NUMBER() OVER(ORDER BY salary DESC) AS row_num
FROM employees;
```

**Output**

|name|salary|row_num|
|----|----|----|
|Charlie|80000|1|
|Frank|80000|2|
|Emma|75000|3|
|Bob|70000|4|
|David|60000|5|
|Alice|50000|6|
|Grace|50000|7|


## 2. RANK vs DENSE_RANK

```python
SELECT name,
salary,
RANK() OVER(ORDER BY salary DESC) AS rank_salary,
DENSE_RANK() OVER(ORDER BY salary DESC) AS dense_rank_salary
FROM employees;
```

**Output**

|name|salary|rank_salary|dense_rank_salary|
|----|----|----|----|
|Charlie|80000|1|1|
|Frank|80000|1|1|
|Emma|75000|3|2|
|Bob|70000|4|3|
|David|60000|5|4|
|Alice|50000|6|5|
|Grace|50000|6|5|


## 3. Top 3 Salaries Per Department

```python
SELECT *
FROM(
 SELECT *,
 ROW_NUMBER() OVER(PARTITION BY department ORDER BY salary DESC) r
 FROM employees
)t
WHERE r<=3;
```

**Output**

|name|department|salary|r|
|----|----|----|----|
|Emma|HR|75000|1|
|Alice|HR|50000|2|
|Grace|HR|50000|3|
|Charlie|IT|80000|1|
|Frank|IT|80000|2|
|Bob|IT|70000|3|
|David|Finance|60000|1|


## 4. Running Total

```python
SELECT emp_id,name,salary,
SUM(salary) OVER(ORDER BY emp_id) AS running_total
FROM employees;
```

**Output**

|name|salary|running_total|
|----|----|----|
|Alice|50000|50000|
|Bob|70000|120000|
|Charlie|80000|200000|
|David|60000|260000|
|Emma|75000|335000|
|Frank|80000|415000|
|Grace|50000|465000|


## 5. Partition By Example (Department Average)

```python
SELECT name,department,salary,
AVG(salary) OVER(PARTITION BY department) AS dept_avg_salary
FROM employees;
```

**Output Example**

|name|department|salary|dept_avg_salary|
|----|----|----|----|
|Alice|HR|50000|58333|
|Emma|HR|75000|58333|
|Grace|HR|50000|58333|
|Bob|IT|70000|76666|
|Charlie|IT|80000|76666|
|Frank|IT|80000|76666|


## 6. Find Duplicate Rows

```python
SELECT *
FROM(
 SELECT *,
 ROW_NUMBER() OVER(PARTITION BY department,salary ORDER BY emp_id) r
 FROM employees
)t
WHERE r>1;
```

**Output**

|name|department|salary|r|
|----|----|----|----|
|Frank|IT|80000|2|
|Grace|HR|50000|2|


## 7. LAG Example

```python
SELECT name,salary,
salary - LAG(salary) OVER(ORDER BY emp_id) AS salary_diff
FROM employees;
```

**Output Example**

|name|salary|salary_diff|
|----|----|----|
|Alice|50000|NULL|
|Bob|70000|20000|
|Charlie|80000|10000|
|David|60000|-20000|


## 8. LEAD Example

```python
SELECT name,salary,
LEAD(salary) OVER(ORDER BY emp_id) - salary AS next_salary_diff
FROM employees;
```

**Output Example**

|name|salary|next_salary_diff|
|----|----|----|
|Alice|50000|20000|
|Bob|70000|10000|
|Charlie|80000|-20000|
|David|60000|15000|
